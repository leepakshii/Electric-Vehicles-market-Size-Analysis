# TASK 2-FIND OUTLIERS IN ENERGY CONSUMPTION

```python
df['mean - Energy consumption [kWh/100 km]'].describe()
```

```python
#Defining quartile 1 and quartile 3
Q1 = df['mean - Energy consumption [kWh/100 km]'].quantile(0.25)
Q3 = df['mean - Energy consumption [kWh/100 km]'].quantile(0.75)
IQR = Q3 - Q1
# Define the bounds
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR #anything falls below this bounds are outliers
#Finding outliers
outliers = df[
    (df['mean - Energy consumption [kWh/100 km]']<lower_bound) | (df['mean - Energy consumption [kWh/100 km]']>upper_bound)
]
outliers[['Car full name','mean - Energy consumption [kWh/100 km]']]
```

```python
from scipy.stats import zscore
```

```python
df['z_score'] = zscore(df['mean - Energy consumption [kWh/100 km]'])
#defining threshold
z_outliers = df[(df['z_score'] > 3) | (df['z_score'] < -3)]
z_outliers[['Car full name','mean - Energy consumption [kWh/100 km]','z_score']]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
sns.boxplot(x=df['mean - Energy consumption [kWh/100 km]'])
plt.title("Energy Consumption Outlier Detection")
plt.show()
```

# Task 4: Build an EV recommendation class. 
The class should allow users to input their
budget, desired range, and battery capacity. The class should then return the top three EVs
matching their criteria

```python
class EVRecommender:
    def __init__(self, dataframe):
        self.df = dataframe.copy()

    def recommend(self, budget, min_range, min_battery):
        # Filter based on user preferences
        filtered = self.df[
            (self.df['Minimal price (gross) [PLN]'] <= budget) &
            (self.df['Range (WLTP) [km]'] >= min_range) &
            (self.df['Battery capacity [kWh]'] >= min_battery)
        ]

        # Sort by range descending and return top 3
        top_matches = filtered.sort_values(by='Range (WLTP) [km]', ascending=False).head(3)

        return top_matches[['Car full name', 'Make', 'Minimal price (gross) [PLN]',
                            'Range (WLTP) [km]', 'Battery capacity [kWh]']]
```

```python
recommender=EVRecommender(df)
recommender.recommend(
 budget=350000,
 min_range=400,
 min_battery=60
)
```

# Taking user input

```python
budget = int(input("Enter your budget (PLN): "))
min_range = int(input("Enter minimum range (km) you want in ur EV : "))
min_battery = int(input("Enter minimum battery capacity (kwh): "))
recommender.recommend(budget, min_range, min_battery)
```

### Task 4: EV Recommendation Class

We designed a class `EVRecommender` that allows users to enter:
- Budget (in PLN)
- Desired minimum range (in km)
- Minimum battery capacity (in kWh)

The system filters EVs based on these preferences and returns the top 3 matching vehicles sorted by highest range.

This class can be reused and embedded in a web app, chatbot, or even customer-facing recommendation engine.


# Task 5: Inferential Statistics – Hypothesis Testing: 
Test whether there is a significant
difference in the average Engine power [KM] of vehicles manufactured by two leading
manufacturers i.e. Tesla and Audi

```python
from scipy.stats import ttest_ind
```

```python
df_cleaned = df.dropna(subset=['Engine power [KM]'])
#Getting the engine power data for tesla and audi
tesla_power = df_cleaned[df_cleaned['Make'] == 'Tesla']['Engine power [KM]']
audi_power = df_cleaned[df_cleaned['Make'] == 'Audi']['Engine power [KM]']
```

```python
t_stat, p_value = ttest_ind(tesla_power, audi_power, equal_var=False)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

### Task 5: Hypothesis Testing – Tesla vs Audi Engine Power

We conducted a **two-sample t-test** to compare the average engine power (in KM) of Tesla and Audi EVs.

#### Hypotheses:
- **Null Hypothesis (H₀):** There is no difference in mean engine power between Tesla and Audi.
- **Alternative Hypothesis (H₁):** There is a significant difference in mean engine power.

#### Results:
- **T-statistic:** 1.79
- **P-value:** 0.1068

#### Conclusion:
Since the *p-value is greater than 0.05*, we **fail to reject the null hypothesis**.

➡️ This suggests that **the observed difference in engine power is not statistically significant** at the 5% significance level.  
In other words, Tesla and Audi have **similar engine performance** in this dataset.
